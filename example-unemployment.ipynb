```sparql
SELECT DISTINCT ?notation ?label ?structure ?unitMeasure ?slice
FROM <https://linked.opendata.swiss/graph/zh/statistics>
WHERE {
  ?subject a <http://purl.org/linked-data/cube#DataSet>;
        <http://www.w3.org/2004/02/skos/core#notation> ?notation;
        <http://www.w3.org/2000/01/rdf-schema#label> ?label;
        <http://purl.org/linked-data/cube#structure> ?structure;
        <http://purl.org/linked-data/cube#slice> ?slice;
        <http://purl.org/linked-data/sdmx/2009/attribute#unitMeasure> ?unitMeasure.
    FILTER REGEX(?label, "Bevölkerung*", "i")
}
LIMIT 10
```

```sparql
# Subjects pointing at a dataset
PREFIX dataset: <https://ld.stadt-zuerich.ch/statistics/dataset/>

SELECT DISTINCT ?type (COUNT(*) AS ?count)
{ 
    ?s ?p dataset:ABL-RAUM-ZEIT-SEX;
       a ?type
}
GROUP BY ?type
ORDER BY DESC(?count)
LIMIT 10
```

```sparql
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX qb: <http://purl.org/linked-data/cube#>

SELECT ?dataset ?notation ?label (COUNT(*) AS ?count) 
FROM <https://linked.opendata.swiss/graph/zh/statistics>
WHERE{
    ?observation a qb:Observation;
        qb:dataSet ?dataset.
    ?dataset rdfs:label ?label;
             <http://www.w3.org/2004/02/skos/core#notation> ?notation;
    FILTER REGEX(?label, "Bevölkerung*", "i")
}
GROUP BY ?dataset ?label ?notation
ORDER BY DESC(?count)
```

```sparql
# Properties in dataset (properties of everything pointing to the dataset)
PREFIX qb: <http://purl.org/linked-data/cube#>
PREFIX dataset: <https://ld.stadt-zuerich.ch/statistics/dataset/>

SELECT DISTINCT ?property
FROM <https://linked.opendata.swiss/graph/zh/statistics>
WHERE {
  ?subject qb:dataSet dataset:BEW-RAUM-ZEIT-ALT-SEX;
             ?property ?object.
}
LIMIT 50
```

```sparql
PREFIX qb: <http://purl.org/linked-data/cube#>
PREFIX dataset: <https://ld.stadt-zuerich.ch/statistics/dataset/>
PREFIX attribute: <https://ld.stadt-zuerich.ch/statistics/attribute/>
PREFIX property: <https://ld.stadt-zuerich.ch/statistics/property/>

SELECT DISTINCT ?time ?space ?hel ?alt ?value
FROM <https://linked.opendata.swiss/graph/zh/statistics>
WHERE {
  ?subject a qb:Observation; 
             qb:dataSet dataset:ABL-RAUM-ZEIT-ALT-HEL;
            <https://ld.stadt-zuerich.ch/statistics/property/ZEIT> ?time;
            <https://ld.stadt-zuerich.ch/statistics/property/RAUM> ?space;
            <https://ld.stadt-zuerich.ch/statistics/property/HEL> ?hel;
            <https://ld.stadt-zuerich.ch/statistics/property/ALT> ?alt;
            <https://ld.stadt-zuerich.ch/statistics/measure/ABL> ?value;
    FILTER (isNumeric(?value))
}
```

```sparql
from SPARQLWrapper import SPARQLWrapper, SPARQLWrapper2, JSON, POST

sparql = SPARQLWrapper2("https://lindas.admin.ch/query")
sparql.setQuery("""
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX qb: <http://purl.org/linked-data/cube#>

    SELECT ?dataset ?label (COUNT(DISTINCT ?observation) AS ?observations)
    FROM <https://linked.opendata.swiss/graph/zh/statistics>
    WHERE{
        ?observation a qb:Observation;
            qb:dataSet ?dataset.
        ?dataset rdfs:label ?label.
    }
    GROUP BY ?dataset ?label
    ORDER BY DESC(?observations)
""")
sparql.setMethod(POST)
sparql.setReturnFormat(JSON)
results = sparql.query().bindings
```
